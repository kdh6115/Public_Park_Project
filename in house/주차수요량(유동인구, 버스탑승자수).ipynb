# 유동인구

## 유동인구 1월

```python
유동인구_1월 = pd.read_csv('유동인구_1월.csv', encoding = 'euc_kr')
유동인구_1월.drop(['time', 'age', 'resd_pop_cnt', 'base_date'], axis = 1, inplace = True)
유동인구_1월 = round(유동인구_1월.groupby(['emd_nm']).sum())
유동인구_1월.rename(columns = {'visit_pop_cnt' : '1월방문객수'}, inplace = True)
유동인구_1월
```

## 유동인구 8월

```python
유동인구_8월 = pd.read_csv('유동인구_8월.csv', encoding = 'euc_kr')
유동인구_8월.drop(['time', 'age', 'resd_pop_cnt', 'base_date'], axis = 1, inplace = True)
유동인구_8월 = round(유동인구_8월.groupby(['emd_nm']).sum())
유동인구_8월.rename(columns = {'visit_pop_cnt' : '8월방문객수'}, inplace = True)
유동인구_8월
```

## 유동인구 데이터

```python
유동인구 = pd.concat([유동인구_1월, 유동인구_8월], axis = 1)
유동인구 = 유동인구.reset_index()
유동인구
```

### emd_nm에서 제주시 빼기

```python
a=유동인구['emd_nm'].str.split(" ")
유동인구['emd_nm']=a.str[1]
유동인구
```

## 평균 유동인구 구하기

```python
유동인구['mean_유동인구'] = ((유동인구['1월방문객수']+유동인구['8월방문객수'])/2)*12
유동인구.set_index(['emd_nm'], inplace = True)
유동인구
```

# 버스탑승자수

```python
종합.rename(columns = {'읍면동' : 'emd_nm'}, inplace = True)
종합.set_index(['emd_nm'], inplace = True)
버스탑승자수 = 종합.iloc[:,-2:]
버스탑승자수
```

### 평균 버스 탑승자수

```python
버스탑승자수['mean_bus'] = ((버스탑승자수['버스탑승자수(1월)']+버스탑승자수['버스탑승자수(8월)'])/2)*12
버스탑승자수
```

# 주차수요량

```python
주차수요량 = 유동인구['mean_유동인구']-버스탑승자수['mean_bus']
주차수요량 = pd.DataFrame(주차수요량, columns = ['주차수요량'])
주차수요량
```

# 데이터 합치기

```python
df = pd.concat([종합, 버스탑승자수['mean_bus'], 유동인구, 주차수요량], axis = 1)
df.drop(['유동인구', '교통량'], axis = 1, inplace = True)
df
```

```python
df.to_csv('주차수요량.csv', encoding = 'euc_kr')
```

```python
df2 = df.reset_index()
df2 = df2.loc[:, ['emd_nm', '주차수요량', 'WiFi 접속자수', '카드이용자수', '자동차등록수', '사업체수', '총인구수', '면수']]
df2
```

```python
df2['2인가구'] = round(df2['주차수요량']/2)
df2['3인가구'] = round(df2['주차수요량']/3)
df2['4인가구'] = round(df2['주차수요량']/4)
df2['5인가구'] = round(df2['주차수요량']/5)
df2.rename(columns = {'주차수요량' : '1인가구'}, inplace = True)
```

```python
df3 = df2.loc[:, ['emd_nm', '1인가구', '2인가구', '3인가구', '4인가구', '5인가구',
            'WiFi 접속자수', '카드이용자수', '자동차등록수', '사업체수', '총인구수', '면수']]
df3
```

```python
df3.to_csv('가구분류.csv', encoding = 'euc_kr')
```
