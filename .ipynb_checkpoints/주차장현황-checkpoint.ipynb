```python
import pandas as pd
```

```python
주차장현황 = pd.read_csv('data/제주도_주차장현황.csv', encoding = 'euc_kr')
주차장현황
```

```python
a = 주차장현황.iloc[:6,:4]
b = 주차장현황.iloc[:6,-2:]
tot = pd.concat([a, b], axis = 1)
tot
```

```python
jeju = tot.iloc[2:4,]
jeju
d = round(int(jeju.iloc[1, 4])/int(jeju.iloc[0, 4]), 2)
e = round(int(jeju.iloc[1, 5])/int(jeju.iloc[0, 5]), 2)

a = ["합계", "제주시", "주차장1개당 평균면수", "평균",d, e]
a = pd.DataFrame(a).T
a.columns = jeju.columns
```

```python
jeju = pd.concat([jeju, a])
jeju.to_csv("제주주차장.csv", encoding = 'euc_kr')
```

```python
seo = tot.iloc[4:6,]

d = round(int(seo.iloc[1, 4])/int(seo.iloc[0, 4]), 2)
e = round(int(seo.iloc[1, 5])/int(seo.iloc[0, 5]), 2)

a = ["합계", "서귀포시", "주차장1개당 평균면수", "평균",d, e]
a = pd.DataFrame(a).T
a.columns = seo.columns
```

```python
seo = pd.concat([seo, a])
seo.to_csv("서귀포주차장.csv", encoding = 'euc_kr')
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(
    fname = 'c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
x = seo.iloc[0,:].index
y = seo.iloc[0,:].values
y= y.astype(int)
plt.title('서귀포 주차장수')
plt.bar(x,y)
plt.show()
```

```python
p = seo.iloc[1,:].index
q = seo.iloc[1, :].values
q = q.astype(int)
plt.title("서귀포 주차면수")
plt.bar(p,q)
plt.show()
```

```python
e=d[d['행정구역(시)별']!='합계']
e.dropna(axis = 0, inplace = True)
e
```

```python
십칠=e.iloc[:,:3]
십칠['2017 년'] = 십칠['2017 년'].astype(int)
십칠
```

```python
십칠2 = 십칠.groupby(['주차장종류별', '행정구역(시)별']).sum().reset_index(drop=False)
십칠2
```

```python
십팔=e.drop(['2017 년'], axis = 1)
십팔['2018 년'] = 십팔['2018 년'].astype(int)
십팔
```
