```python
import pandas as pd
```

```python
df1 = pd.read_csv('data/유동인구_1월.csv', encoding = 'euc_kr')
df2 = pd.read_csv('data/유동인구_8월.csv', encoding = 'euc_kr')
```

```python
df1
```

```python
a = df1.copy()
b = a.drop(['time', 'age', 'resd_pop_cnt'], axis = 1)
c = round(b.groupby(['base_date','emd_nm']).mean())
c.rename(columns = {'visit_pop_cnt' : '1월 유동인구'}, inplace=  True)
d = c.groupby(['emd_nm']).sum()
d['1월 유동인구'] = d['1월 유동인구'] * 30
d
```

```python
e = df2.copy()
f = e.drop(['time', 'age', 'resd_pop_cnt'], axis = 1)
f = round(f.groupby(['base_date','emd_nm']).mean())
f.rename(columns = {'visit_pop_cnt' : '8월 유동인구'}, inplace=  True)
g = f.groupby(['emd_nm']).sum()
g['8월 유동인구'] = g['8월 유동인구'] * 30
g
```

```python
df3 = pd.concat([d, g], axis = 1)
df3
```

```python
import numpy as np
df3['mean_유동인구'] = round((df3['1월 유동인구']+df3['8월 유동인구'])/2)
df3
```

```python
유동인구 = df3.loc[:,'mean_유동인구']
유동인구
```

```python
유동인구.to_csv('유동인구.csv', encoding = 'euc_kr')
```
