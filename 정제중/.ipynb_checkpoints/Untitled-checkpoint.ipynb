```python
import pandas as pd
```

```python
df = pd.read_excel('유동인구1월8월.xlsx')
df.rename(columns = {'Unnamed: 0' : '주소'}, inplace = True)
df.drop(['1_resd_pop_cnt', '8_resd_pop_cnt'], axis = 1, inplace = True)
df
```

```python
df2 = df.copy()
df2['mean_visit_pop_cnt_12'] = ((df2['1_visit_pop_cnt']+df2['8_visit_pop_cnt'])/2)*12
df2['mean_visit_pop_cnt_1'] = round((df2['1_visit_pop_cnt']+df2['8_visit_pop_cnt'])/2)
df2.drop(['1_visit_pop_cnt', '8_visit_pop_cnt'], axis = 1, inplace = True)
df2
```

```python
a = pd.read_csv('유동인구_1월.csv', encoding = 'euc_kr')
b = pd.read_csv('유동인구_8월.csv', encoding = 'euc_kr')
```

```python
a.drop(['time', 'age', 'resd_pop_cnt', 'base_date'], axis = 1, inplace = True)
b.drop(['time', 'age','resd_pop_cnt', 'base_date'], axis = 1, inplace = True)
```

```python
a = a.groupby(['emd_nm']).sum()
a.rename(columns = {'visit_pop_cnt' : '1월방문객수'}, inplace = True)
a
```

```python
b = b.groupby(['emd_nm']).sum()
b.rename(columns = {'visit_pop_cnt' : '8월방문객수'}, inplace = True)
b
```

```python
tot = pd.concat([a, b['8월방문객수']], axis = 1)
tot
```

```python
tot.to_csv('1_8유동인구_x24.csv', encoding = 'euc_kr')
```
