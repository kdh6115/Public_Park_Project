```python
park = split.apply(lambda x : pd.Series(x))
park.rename(columns = {0 : '주소', 1:'lat', 2 : 'loc'}, inplace = True)
park
```

```python
park.to_csv('data/공영주차장좌표.csv', header = True, encoding = 'EUC-KR')
```

```python
loc = park.iloc[:,1:]
loc
```

```python
address = park['주소']
address
```

```python
import matplotlib.pyplot as plt
import folium
map_osm = folium.Map(location=[33.362634, 126.529167], zoom_start=10)
map_osm
```

> 팝업 한글 깨지는데 이건 map.html에서는 안 깨지기 때문에 그거 참고

```python
from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(map_osm)

for i in range(25):
  folium.Marker(
    [loc.iloc[i:i+1,0], loc.iloc[i:i+1,1]],
      popup = address[i],
    icon=folium.Icon(color='blue',icon='ok'),
  ).add_to(marker_cluster)
map_osm.save("map.html")
map_osm
```
